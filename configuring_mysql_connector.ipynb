```python
# Query para obtener detalles de clientes con órdenes mayores a $60
high_value_query = """
SELECT 
    c.FullName,
    c.Phone,
    c.Email,
    o.TotalCost
FROM Customers AS c
JOIN Orders AS o
    ON c.CustomersID = o.fk_orders_customer
WHERE o.TotalCost > 60;
"""

cursor.execute(high_value_query)

results = cursor.fetchall()

for row in results:
    print(row)
```
